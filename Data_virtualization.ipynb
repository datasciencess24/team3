```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np
```

```python
### get the path of the files
dic_OK = './Data/OK_Measurements'
dic_NOK = './Data/NOK_Measurements'

file_AE = 'raw/Sampling2000KHz_AEKi-0.parquet'
file_AC = 'raw/Sampling100KHz_Irms_Grinding-Grinding spindle current L1-Grinding spindle current L2-Grinding spindle current L3-0.parquet'

folder_OK= os.listdir(dic_OK)
folder_NOK= os.listdir(dic_NOK)
folder_OK.remove('.DS_Store')
folder_NOK.remove('.DS_Store')

print('folder_OK: ', folder_OK)
print('folder_NOK: ', folder_NOK)
```

```python
######## open the files ######
dataframe_OK_AE = []
for file_name in folder_OK:
    file_path = os.path.join(dic_OK, file_name,file_AE)
    if os.path.isfile(file_path):
        df = pd.read_parquet(file_path,engine='fastparquet')
        dataframe_OK_AE.append(df)
```

```python
dataframe_NOK_AE = []
for file_name in folder_NOK:
    file_path = os.path.join(dic_NOK, file_name,file_AE)
    if os.path.isfile(file_path):
        df = pd.read_parquet(file_path,engine='fastparquet')
        dataframe_NOK_AE.append(df)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(12, 6))
print(axes.shape)
for i in range(axes.shape[0]):
    for j in range(axes.shape[1]):
        if i == 0:
            axes[i,j].plot(dataframe_OK_AE[j*2])
            axes[i,j].set_title("OK_signal")
        else:
            axes[i,j].plot(dataframe_NOK_AE[j*2])
            axes[i,j].set_title("NOK_signal")
```

```python
###### open the files for the current#
dataframe_OK_AC = []
for file_name in folder_OK:
    file_path = os.path.join(dic_OK, file_name,file_AC)
    if os.path.isfile(file_path):
        df = pd.read_parquet(file_path,engine='fastparquet')
        dataframe_OK_AC.append(df)
```

```python
dataframe_NOK_AC = []
for file_name in folder_NOK:
    file_path = os.path.join(dic_NOK, file_name,file_AC)
    if os.path.isfile(file_path):
        df = pd.read_parquet(file_path,engine='fastparquet')
        dataframe_NOK_AC.append(df)
```
